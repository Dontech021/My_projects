#### Determinant
The function determinant can solve for the determinant of squared matrix of any order.

```python
def determinant(matrix):     #determinant function
    def collect_matrix(mat,r,c):                                    #The collect_matrix function inside determinant function helps to remove the row,r and column,c
        new_mat=mat                                                 #from a matrix,matrix.
        new_mat=np.delete(new_mat,r,axis=0)
        new_mat=np.delete(new_mat,c,axis=1)
        return new_mat

    def modulo1(mat):                                               #The function modulo1 uses the funtion(degenerator) within it called degen to yield the product of
        def degen(matgen):                                          #each first row element,-1 raised to the power of index of the element and the modulo2 of the collect_matrix of
            for m in range(matgen.shape[1]):                        #the matrix over the column and row of the element. it continues to yield this product for all first row elements
                yield ((-1)**m)*matgen[0][m]*modulo2(collect_matrix(matgen,0,m))#and sum their products up. The two modulo1 and modulo2 works the same way and they call each other 
                                                                    #until the collected matrix becomes 2x2 matrix. Hence, the determinant function method of solving mathrix is 
        if mat.shape==(2,2):                                        #similar to the normal way of solving matrix
            det=mat[0][0]*mat[1][1]-mat[0][1]*mat[1][0]
            return det
        else:
            adds=0
            for dmat in degen(mat):
                adds+=dmat

            return adds

    def modulo2(mat):                                               #The function modulo2 uses the funtion(degenerator) within it called degen to yield the product of
        def degen(matgen):                                          #each first row element,-1 raised to the power of index of the element and the modulo1 of the collect_matrix of
            for m in range(matgen.shape[1]):                        #the matrix over the column and row of the element. it continues to yield this product for all first row elements
                yield ((-1)**m)*matgen[0][m]*modulo1(collect_matrix(matgen,0,m))#and sum their products up. The two modulo1 and modulo2 works the same way and they call each other 
                                                                    #until the collected matrix becomes 2x2 matrix. Hence, the determinant function method of solving mathrix is 
        if mat.shape==(2,2):                                        #similar to the normal way of solving matrix
            det=mat[0][0]*mat[1][1]-mat[0][1]*mat[1][0]
            return det
        else:
            adds=0
            for dmat in degen(mat):
                adds+=dmat

            return adds        
    if matrix.shape[0] != matrix.shape[1] or matrix.shape[0]==1:
        return f'A matrix of order {matrix.shape} has no determinant'
    else:
        determinant=modulo1(matrix)                                 #The function determinant uses the function modulo to solve for the determinant of a matrix
        return determinant
```

#### Cofactor
The function cofactor can solve for the cofactor of squared matrix of any order.

$X =

#### Adjoint
The function adjoint can solve for the adjoint of squared matrix of any order.

```python
def adjoint(matrix):                                                                         #adjoint
    return transpose(cofactor(matrix))                              #finds the transpose of the cofactor of a matrix
    
```

#### Trace
The function trace can solve for the trace of squared matrix of any order.

```python
def trace(matrix):
    if matrix.shape[0] != matrix.shape[1] or matrix.shape[0]==1:
        return f'A matrix of order {order} has no trace'
    trace=0
    for i in range(matrix.shape[0]):
        trace+=matrix[i][i]                                        
    return trace                                                                   #it returns the trace of the matrix which is the sum of the leading diagonals
```
