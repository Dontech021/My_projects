# Milestone Project 1: BUILD A TIC TAC TOE GAME


```python
from IPython.display import clear_output

continue_game='Y'
while continue_game=='Y': 
    stop=0
    place_values=[1,2,3,4,5,6,7,8,9]
    [A,B,C,D,E,F,G,H,I]=place_values

    #tic tac toe board displayer
    def board():
        print('','- '*16,'\n','|         '*4,
              '\n','|   ',A,'   |   ',B,'   |   ',C,'   |\n','|         '*4,
              '\n','- '*16,'\n','|         '*4,
              '\n','|   ',D,'   |   ',E,'   |   ',F,'   |\n','|         '*4,
              '\n','- '*16,'\n','|         '*4,
              '\n','|   ',G,'   |   ',H,'   |   ',I,'   |\n','|         '*4,
              '\n','- '*16,'\n')
    #function that collects position choice from player
    def choice(player):
        place_value=[A,B,C,D,E,F,G,H,I]

        playerplace=int(input(player+' Enter position: '))
        while playerplace not in place_value:
            print(player+' please choose a position from the board above')
            playerplace=int(input(player+' Enter position: '))   
        return playerplace   

    #winner checker functinon  
    def check():               
        if A==B==C=='X' or D==E==F=='X'or G==H==I=='X' or A==E==I=='X' or G==E==C=='X' or A==D==G=='X' or E==B==H=='X' or I==F==C=='X':
            board()
            print('CONGRATULATIONS TO player1. YOU HAVE WON THE GAME')
            return 1
        elif A==B==C=='O' or D==E==F=='O'or G==H==I=='O' or A==E==I=='O' or G==E==C=='O' or A==D==G=='O' or E==B==H=='O' or I==F==C=='O':
            board()
            print('CONGRATULATIONS TO player2. YOU HAVE WON THE GAME')
            return 1
        else:
            return 0

    #FUNTION FOR PHASE TWO OF THE GAME
    #ASKS PLAYER THE SEED TO MOVE AND MAKES SURE HE SELECTS IT CORRECTLY
    def player_choice_lvl2(count):

        choices=[1,2,3,4,5,6,7,8,9]
        choice=10
        while count%2==0:
            choice=int(input('player1 choose the position of the seed to move: '))
            while choice not in choices:
                print('player1 please choose a position from the initial board')
                choice=int(input('Enter position: '))
            if place_values[choice-1]!='X':
                print('Wrong position. Please player1 choose again')
            elif place_values[choice-1]=='X':
                return choice
        while count%2!=0:
            choice=int(input('player2 choose the position of the seed to move: '))
            while choice not in choices:
                print('player2 please choose a position from the initial board')
                choice=int(input('Enter position: '))
            if place_values[choice-1]!='O':
                print('Wrong position. Please player2 choose again')
            elif place_values[choice-1]=='O':
                return choice

    count1=0
    count2=0

    print('please make a choice of what player you want to be. \nIs it player1 or player2?')
    print('player1 has \'X\' while player2 has \'O\'')
    print('Here comes the board:\n\n')
    while count1<8 and (A!=B or B!=C) and (D!=E or E!=F) and (G!=H or H!=I) and (A!=E or E!=I) and (G!=E or E!=C) and (A!=D or D!=G) and (B!=E or E!=H) and (C!=F or F!=I) and (D!=E or E!=F):
        if count1%2==0:
            board()
            #ask player1 position and replace with '*''
            player1choice=choice('player1')
            place_values.insert(place_values.index(player1choice),'X')
            place_values.remove(player1choice)        
            count1+=1
            count2+=1

        elif count2==8:
            break     

        elif count2%2!=0:

            board()
            #ask player2 position and replace with '#'
            player2choice=choice('player2')
            place_values.insert(place_values.index(player2choice),'O')
            place_values.remove(player2choice)
            count1+=1 
            count2+=1

        [A,B,C,D,E,F,G,H,I]=place_values
        clear_output()
    clear_output()
    stop=check()
    A,B,C,D,E,F,G,H,I=1,2,3,4,5,6,7,8,9
    
    if stop==0:
        print('this game is getting tougher. You guys are really pros')
        print('Ok, now players has to choose which of their seed they want to move.')
        print('each player has to choose position only to move their seed that is in that position to the emnty space.')
        print('use the initial board label which is below +to remember your board position')
        board()
        input('Press any key to continue if you understand: ')
        clear_output()

    counted1=0
    counted2=0
    count=0
    while stop==0:         
        [A,B,C,D,E,F,G,H,I]=place_values
        board()
        position=player_choice_lvl2(count)
        while counted1%2==0:
            players_seed=['O','X']
            for value in place_values:
                if value not in players_seed:
                    value_position=place_values.index(value)
                    place_values.remove(value)                
                    place_values.insert(value_position,'X') 
                    place_values.pop(position-1)               
                    place_values.insert(position-1,position)
                    clear_output()
                    counted1+=1
                    count+=1
                    break

        while counted2%2!=0:
            players_seed=['O','X']
            for value in place_values:
                if value not in players_seed:
                    value_position=place_values.index(value)
                    place_values.remove(value)                
                    place_values.insert(value_position,'O') 
                    place_values.pop(position-1)               
                    place_values.insert(position-1,position)
                    clear_output()
                    counted2+=1
                    count+=1
                    break
        [A,B,C,D,E,F,G,H,I]=place_values
            
        stop=check()
        if counted2%2==0:
            counted2+=1
        elif counted1%2!=0:
            counted1+=1
    while True:
        continue_choice=['Y','N','n','y']
        continue_game=input('Do you still want to continue the game?\n\nPress Y to continue\n\nOr press N to stop playing: ').upper()
        if continue_game not in continue_choice:
            clear_output()
            print('Please choose Y or N')
            continue
        elif continue_game=='Y' or continue_game=='N':
            if continue_game=='Y':
                clear_output()
                print('I\'ll take Y as YES')
            elif continue_game=='n':
                clear_output()
                print('I\'ll take N as NO')         
            break
                
```
