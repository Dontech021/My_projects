```python
#import neccessary libraries
from IPython.display import clear_output
import random
```

```python
#create suits, ranks and assign values to the ranks
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}
```

### Create some Object like Card, Deck, Player,Bet 

```python
#class object for class
class Card:
    
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[rank]
        
    def print_card(self):
        card=self.rank+' of '+self.suit
        print(f'       {card:<20}' ,end='')
        
    def __str__(self):
        return (self.rank+' of '+self.suit)
```

```python
#class object for deck
class Deck:
    
    def __init__(self):
        self.cards=[]
        self.ace=[]
        for suit in suits:
            for rank in ranks:
                self.cards.append(Card(suit,rank))
        for suit in suits:
            self.ace.append(Card(suit,ranks[12]))
                
    def deck_shuffle(self):
        random.shuffle(self.cards)
    
    def one_out(self):
        return self.cards.pop()

    
```

```python
#class object for players
class Player:
    
    def __init__(self,name,account):
        self.cards=[]
        self.name=name
        self.account=account
        
    def withdraw(self,amount):
        self.account-=amount

    def deposit(self,amount):
        self.account+=amount
    
    def __str__(self):
        return (self.name+f" balance: ${self.account}")
    

```

```python
#class object for bet
class Bet:
    
    def __init__(self,amount, All):
        self.amount=amount
        self.All=All

    def __str__(self):
        return (f" Bet: ${self.amount}                           Balance: ${self.All-self.amount}")
    def add(self,amount):
        self.amount+=amount
    def clear(self):
        self.amount=0
    def less_fund(self,amount):
        clear_output()
        print(f'Insufficient balancese, Bet: ${self.amount+amount}')
```

```python
#Funtion that asks player to choose bet amount or do other things like clear, all or deal(which means to use the current amount)
def player_bet_choices(player):
    print(f'{player}: \'A\' or \'C\' or \'D\' or bet amount:',end='')
    choice= input()
    try:
        int(choice)
        return int(choice) 
    except:
        pass
    choice = choice.upper() 
    while choice not in ['A','D','C']:
        print(f'{player} Choose either All, Clear or Deal or bet amount')
        print('\'A\' or \'C\' or \'D\' or integer:',end='')
        choice= input()

        try:
            int(choice)
            return int(choice) 
        except:
            pass
        choice = choice.upper()
    return choice 
```

```python
#funtion allows players to select the amount they want to bet with
#and it will return a Bet object
def bet(player,amount,All):
    player_bet=Bet(amount, All)          #creates a new Bet object
    print(player_bet)
    print('press \'D\' if you wish to deal or \'C\' to clear or just pick betting amount \n')
    deal=player_bet_choices(player)      #asks user to make their choice on bet amount
    
    while True:
        if deal=='A':
            clear_output()
            player_bet=Bet(All,All)                                                  #selects the players available balance
            print(player_bet)                                                        #and wait for confirmation from player
            print('press \'D\' if you wish to deal or \'C\' to clear or just pick betting amount \n')
            deal=player_bet_choices(player)
            if deal=='D':
                clear_output()
                if player_bet.amount<5:
                    print ('Bet amount is too small')
                    deal=player_bet_choices(player)
                    continue
                return player_bet                                                      #returns player_bet if player choose D
            elif deal=='A':
                continue
            elif deal=='C':
                clear_output()
                player_bet.clear()
                print(player_bet)
                print(f'\n{player} pick betting amount: ')
                deal=player_bet_choices(player)
                continue
            else:
                continue

        else:
            if deal=='D':
                clear_output()
                if player_bet.amount<5:
                    print ('Bet amount is too small')
                    deal=player_bet_choices(player)
                    continue
                return player_bet                                                           #return  player_bet
            elif deal=='C':
                clear_output()
                player_bet.clear()
                print(player_bet)
                print(f'\n{player} pick betting amount: ')
                deal=player_bet_choices(player)
                continue
            elif int(deal)+player_bet.amount>All:                           #checks if the player has selected more than he/she has
                player_bet.less_fund(int(deal))                 
                player_bet=Bet(All,All)
                print(player_bet)
                print('press \'D\' if you wish to deal or \'C\' to clear or just pick betting amount \n')
                deal=player_bet_choices(player)
                if deal=='C':
                    player_bet.clear()
                    print(player_bet)
                    print(f'\n{player} pick betting amount: ')
                    deal=player_bet_choices(player)
                    continue
                elif deal =='D':
                    continue
                elif deal =='A':
                    continue
                else:
                    player_bet.clear()
                    continue
            clear_output()
            player_bet.add(int(deal))
            print(player_bet)
            print('press \'D\' if you wish to deal or \'C\' to clear or just pick betting amount \n')
            deal=player_bet_choices(player)

```

```python

hs=['H','S'] 
hssp=['H','S','SP']
hsd=['H','S','D']  
hsspd=['H','S','SP','D'] 

#function to sum value of cards
def cards_sum(cards):
    value = 0
    for card in cards:
        value+=card.value
    return value
#display card openly for both dealer and player
def display_card(owner,cards):
    print(f'\n {owner}\n cards:')
    for card in cards:
        print(f'       {card}') 
    
            
#display card openly for both dealer and player with thier value-sum
def display_cardall(owner,cards,val):
    print(f'{owner} \n cards:')
    for card in cards:
        print(f'       {card}') 
    
    print(f'( {val} )')
            
                

#dealer card displayer 
def dealer_card_dis(cards):
    print(f'Dealer\'s cards:')
    print(f'       {cards[0]}')
    print(f'            ???? \n') 
    

#functions to hit or stand or split, or double 
def hitorstand():
    print('\'H\' or \'S\': ',end='')
    choice= input().upper()
    while choice not in hs:
        print('Choose either hit or stand')
        print('\'H\' or \'S\': ',end='')
        choice= input().upper()
    return choice
def hitorsplitorstand():
    print('\'H\' or \'SP\' or \'S\': ',end='')
    choice= input().upper()
    while choice not in hssp:
        print('Choose either hit or split or stand')
        print('\'H\' or \'Sp\' or \'S\': ',end='')
        choice= input().upper()
    return choice
def hitordoubleorstand():
    print('\'H\' or \'D\' or \'S\': ',end='')
    choice= input().upper()
    while choice not in hsd:
        print('Choose either hit or split or stand')
        print('\'H\' or \'D\' or \'S\': ',end='')
        choice= input().upper()
    return choice
def hitordoubleorsplitorstand():
    print('\'H\' or \'D\' or \'Sp\' or \'S\': ',end='')
    choice= input().upper()
    while choice not in hsd:
        print('Choose either hit or split or stand')
        print('\'H\' or \'D\' or \'Sp\' or \'S\': ',end='')
        choice= input().upper()
    return choice

#function to check if there is a party with card-sum=21
def check_values21(player_val,dealerv):
    player=player_val
    dealer=cards_sum(dealerv)
    if dealer==player==21:
        return 'P'
    elif player==21:
        return 'BJ'
    elif dealer==21:
        return 'D'
    else:
        return False

```

```python
#class object for defining the cards in the hands of the player during play time
class Player_cards:
    def __init__(self):
        self.cards=[]
        self.split1=[]
        self.split2=[]
        self.value=0
    def add(self,card):
        self.cards.append(card)

    def split(self):
        for i in range(len(self.cards)):
            sp1=[self.cards[x] for x in range(i)]
            sp2=[self.cards[x] for x in range(i,len(self.cards))]
            if cards_sum(sp1)==cards_sum(self.cards)/2:
                self.split1=sp1
                self.split2=sp2
                break 
    
    def checker(self):
        ace_num=0
        value=cards_sum(self.cards)
        val=value
        for card in self.cards:
            if card.value==11:        #check how many aces are available
                ace_num+=1
        if ace_num==0:
            print(f'( {value} )')
            self.value=value
        elif ace_num==1:
            if val==21:
                print(f'( {val} )')
                self.value=val
            elif val>21:
                val=value-10
                print(f'( {val} )')
                self.value=val
            else:
                print(f'( {value-10}/{value} )')
                self.value=value

        else:
            val=value-(10*ace_num)
            if val+10==21:
                print(f'( {val+10} )')
                self.value=val+10
            elif val+10<21:
                print(f'( {val}/{val+10} )') 
                self.value= val+10
            else:
                print(f'( {val} )')
                self.value= val
```

```python
           
#class object for defining the cards in the hands of the dealer during play time
class Dealer_cards:
    def __init__(self):
        self.cards=[]
        self.value=0
    def add(self,card):
        self.cards.append(card)

    def checker(self):
        ace_num=0
        value=cards_sum(self.cards)
        val=value
        for card in self.cards:
            if card.value==11:        #check how many aces are available
                ace_num+=1
        if ace_num==0:
            print(f'( {value} )')
            self.value=value
        elif ace_num==1:
            if val==21:
                print(f'( {val} )')
                self.value=val
            elif val>21:
                val=value-10
                print(f'( {val} )')
                self.value=val
            else:
                print(f'( {value-10}/{value} )')
                self.value=value

        else:
            val=value-(10*ace_num)
            if val+10==21:
                print(f'( {val+10} )')
                self.value=val+10
            elif val+10<21:
                print(f'( {val}/{val+10} )') 
                self.value= val+10
            else:
                print(f'( {val} )')
                self.value= val
```

```python
#function to ask player to starts game
#this funtion will return B, D, BJ, W, P,[bust,dealer wins,blackjack,won,push] OR SP for splt

def start_game(deck,player,stake):
    player_cards=Player_cards()
    dealer_cards=Dealer_cards()
    dealer='Dealer'
    for a in range(0,2):
        dealer_cards.add(deck.one_out())              #first four cards shared
        player_cards.add(deck.one_out())
        

    choice=0
    act=0
    while True: 
        if choice=="H":
            dealer_card_dis(dealer_cards.cards)                                          
            print('\n')
            player_cards.add(deck.one_out())
            display_card(player,player_cards.cards)
            player_cards.checker()
            check21=check_values21(player_cards.value,dealer_cards.cards)
            if check21!=False:
                clear_output()
                display_card(dealer,dealer_cards.cards)
                dealer_cards.checker()
                display_cardall(player,player_cards.cards,player_cards.value)
                return check21
            if player_cards.value>21:                                                #returns B as busted
                return 'B'
            choice=0
            act=0
            clear_output()

        elif choice=="S":
            while True:
                clear_output()
                display_cardall(player,player_cards.cards,player_cards.value)
                print('\n')            
                display_card('dealer',dealer_cards.cards)  
                dealer_cards.checker()
                if dealer_cards.value>17:
                    if 21>=dealer_cards.value>player_cards.value:
                        return 'D'
                    elif dealer_cards.value>21:
                        return 'W'
                    elif dealer_cards.value==player_cards.value:
                        return 'P'
                dealer_cards.add(deck.one_out())

    #for stand
        if act==1:
            choice=hitorstand()
            clear_output()
            act=0
            continue
        dealer_card_dis(dealer_cards.cards)                                          
        print('\n')
        display_card(player,player_cards.cards)
        player_cards.checker()
        check21=check_values21(player_cards.value,dealer_cards.cards)
        if check21!=False:
            clear_output()
            display_card(dealer,dealer_cards.cards)
            dealer_cards.checker()
            display_cardall(player,player_cards.cards,player_cards.value)
            return check21
        elif cards_sum(player_cards.cards) <13 and act!=1:             #if double option
            choice=hitordoubleorstand()
            clear_output()

            if choice=='D':
                if stake.amount*2>stake.All:
                    stake.less_fund(stake.amount)
                    act=1
                else:
                    stake.add(stake.amount)
                    print(stake)
                    act=1
                dealer_card_dis(dealer_cards.cards)                                          
                print('\n')
                display_card(player,player_cards.cards)
                player_cards.checker()   
                
            else:
                continue
        act=1

        
```

```python
#function that asks player if they want to change their stake amount or continue game
def start_select(initial):
    print('\n',initial,'\nYES(Continue game) or NO(change stake amount)? \n',end='')
    choice= input().upper()
    while choice not in ["YES",'NO','Y','N']:
        print('Choose either YES(N) or NO(N)')
        choice= input().upper()
    return choice
```
